```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Wine.csv")
data
```

```python
data.columns = data.iloc[1]
data = data[2:]
data
```

```python
data = data.drop(data.columns[:1], axis=1)
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_cols = [
    'Genotypes', 'Replication', 'PruningBiomass', 'TimeofBudbrust',
    'Timeoffullbloom', 'Time of verasion', 'ABW', 'BW', 
    'NOB', 'BD', 'BL', 'TSS', 'Acidity', 
    'Compaactness', 'Skinfirmness', 'Bunlen', 
    'Bunwidth', 'pedunclelength', 'Physiologicalmaturity', 
    'Seed weight ', 'NOS'
]

# Encode each categorical column
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col].astype(str))

# Display the transformed DataFrame
print(data.head())
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(data.corr(),annot=True)
```

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA().fit(data)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Raxtio')
plt.grid()
plt.show()


```

```python

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.axhline(y=0.95, color='r', linestyle='--')  # 95% threshold
plt.show()
```

```python
kmeans = KMeans(n_clusters=11, random_state=42)
clusters = kmeans.fit_predict(principal_components)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Calculate the distance from each point to its cluster centroid
distances = np.linalg.norm(principal_components - centroids[clusters], axis=1)

# Create a DataFrame with the principal components and cluster information
pca_df = pd.DataFrame(principal_components, columns=[f'PC{i+1}' for i in range(principal_components.shape[1])])
pca_df['Cluster'] = clusters
pca_df['Distance_to_Centroid'] = distances

# Calculate cluster summary
cluster_summary = pca_df.groupby('Cluster').agg(
    Sample_Count=('Cluster', 'size'),
    Mean_PC1=('PC1', 'mean'),
    Mean_PC2=('PC2', 'mean'),
    Avg_Distance=('Distance_to_Centroid', 'mean')
).reset_index()

# Display the cluster summary
print(cluster_summary)

# Optional: Visualize the clusters using PC1 and PC2
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster'], cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering on PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid()
plt.legend()
plt.show()
```
